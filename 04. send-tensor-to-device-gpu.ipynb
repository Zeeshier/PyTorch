# Send Tensor to Device/GPU

```python
import torch
```

```python
a = torch.rand(2,3)
```

```python
a
```

## Device Function 

```python
a.device
```

## Check Cuda

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
device
```

## Send Tensor to Cuda using (To Function)

```python
a=a.to(device)
```

```python
a.device
```
