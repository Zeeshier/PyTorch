# Build Linear Regression Model

# Make own Model

```python
import torch
import torch.nn as nn
```

```python
x = torch.tensor([[1.],[2.],[3.],[4.]], dtype = torch.float32)
y = torch.tensor([[2.],[4.],[6.],[8.]], dtype = torch.float32)
```

```python
w = torch.tensor(0.0,requires_grad=True, dtype = torch.float32)
```

## Forward Pass

```python
def forward(x):
    return w*x
```

## Loss Compute

```python
ls = nn.MSELoss()
learning_rate =0.01
optimzer = torch.optim.SGD([w], lr = learning_rate)
```

## Backward

```python
num_epochs = 10

for epoch in range(num_epochs):
    y_pred = forward(x)
    loss = ls(y_pred, y)

    loss.backward()

    optimzer.step()
    optimzer.zero_grad()

    print(f'epoch {epoch}, loss = {loss}')
    
```

# Custom Model


```python
import torch 
import torch.nn as nn
```

```python
x = torch.tensor([[1.],[2.],[3.],[4.]], dtype = torch.float32) ## it must be 2d array, number of rows are number of sample
y = torch.tensor([[2.],[4.],[6.],[8.]], dtype = torch.float32)
```

```python
n_samples, n_features = x.shape
```

```python
input_size = n_features
output_size = n_features
```

```python
lr = 0.01
epochs = 10
```

```python
class Linearregression(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(Linearregression, self).__init__()
        #define layer
        self.linear = nn.Linear(input_dim, output_dim)
    def forward(self, x):
        return self.linear(x)
```

```python
model = Linearregression(input_size, output_size)
```

```python
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=lr)
```

```python
for epoch in range(epochs):
    y_pred = model(x)
    
    l = loss(y_pred, y)
    
    l.backward()
    
    optimizer.step()
    
    optimizer.zero_grad()
    [w,b]= model.parameters()
    print(f'epoch {epoch}, loss= {l}, weight = {w[0].item()}, bias={b[0].item()}')
```
